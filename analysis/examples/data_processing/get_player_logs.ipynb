```python
from nba.players import get_player_game_log, get_player_id

import nba.constants  as c

import os

ROOT_DIR = os.getcwd()[:-33] # verify this points to the root directory of the project

season = c.SEASON_2022_2023
player = c.TYRESE_MAXEY

ROOT_DIR
```

```python
player_id = get_player_id(player)
log = get_player_game_log(player_id, season)
log
```

```python
log.to_csv(ROOT_DIR + '/data/game_logs/{season}/player/{player}_{season}_game_log.csv'.format(player=player, season=season), index=False)
```
